```python
import pandas as pd
import plotly.express as px
from vioPlot import getData
from tqdm import tqdm
data = getData()
data.head()
```

```python
data['review rate number'].unique()
```

```python
scatterFrame = pd.DataFrame()
for neighbourhood in ['Brooklyn', 'Bronx']:
    scatterFrame = pd.concat(
            [scatterFrame,
                data[data["neighbourhood group"] == neighbourhood]
                ]
            )
scatter = px.violin(scatterFrame,
        y = "number of reviews",
        x = "review rate number",
        color = "neighbourhood group",
        box = True)
```

```python
scatter.show()
```
